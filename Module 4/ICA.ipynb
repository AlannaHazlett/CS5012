```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_edge(776, 435)
G.add_edge(776, 869)
G.add_edge(435, 141)
G.add_edge(435, 700)
G.add_edge(141, 66)
G.add_edge(66, 48)
G.add_edge(66, 124)
G.add_edge(48, 42)
G.add_edge(700, 450)
G.add_edge(869, 791)
G.add_edge(791, 806)
G.add_edge(869, 957)
```

```python
# explicitly set positions 
pos = {776: (0, 0), 435: (-8, -4), 141: (-12, -8), 66: (-16, -12), 48: (-20, -16), 42:(-24,-20), 700: (-4, -8), 450: (-8, -12), 124: (-12, -16), 869: (8, -4), 791: (4, -8), 806: (8, -12), 957: (12, -8)}

options = {
    "font_size": 10,
    "node_size": 1000,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 5,
    "width": 5,
}
```

```python
nx.draw_networkx(G, pos, **options)
# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

**In-order Traversal:** 42, 48, 66, 124, 141, 435, 450, 700, 776, 791, 806, 869, 957

**Pre-order Traversal:** 776, 435, 141, 66, 48, 42, 124, 700, 450, 869, 791, 806, 957

**Post-order Traversal:** 42, 48, 124, 66, 141, 450, 700, 435, 806, 791, 957, 869, 776
